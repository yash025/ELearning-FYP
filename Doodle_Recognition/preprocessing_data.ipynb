```python
import numpy
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import csv
import os
import random
import sklearn
from sklearn.model_selection import train_test_split
```

```python
DATA_DIR = "C:/Users/Yashwanth/Downloads/Datasets/Doodle_dataset"
IMG_SIZE = 28
label = ["apple", "arm", "banana", "bandage", "baseball", "bat", "blackberry", "bread", "candle", "carrot", "circle", "cloud", "diamond", "door", "elbow", "feather", "foot", "frying pan", "hat", "hexagon", "hockey stick", "key", "knife", "leaf", "octagon", "paintbrush", "star", "triangle", "zigzag"]
```

```python
def chunks(l):
    return [l[i:i+28] for i in range(0, len(l), 28)]
```

```python
# For Testing the images
def display_image(img):
    img = numpy.array(img, dtype=numpy.uint8)
    plt.imshow(img,cmap = "gray")
    plt.show()
```

```python
def preprocessing_data():
    train_data = []
    i = 0
    for object_data in os.listdir(DATA_DIR):
        path = "C:/Users/Yashwanth/Downloads/Datasets/Doodle_dataset/"
        path += object_data
        data = np.load(path, allow_pickle=True)
        print(label[i])
        j = 0
        for img in tqdm(data):
            img = chunks(img)
            img = numpy.array(img, dtype=np.uint8)
            train_data.append([np.array(img),np.array(label[i])])
            if(j == 0):
                print(img.shape)
                display_image(img)
                j += 1
        i += 1
    return train_data
```

```python
data = preprocessing_data()
print(len(data))
```

```python
random.shuffle(data)
train_data, test_data = sklearn.model_selection.train_test_split(data, train_size = 0.7, test_size = 0.3)
print(len(train_data))
print(len(test_data))
```

```python
#Saving of train data to avoid doing preprocessing again and again
numpy.save("C:/Users/Yashwanth/Downloads/Datasets/train_data.npy",train_data) 
numpy.save("C:/Users/Yashwanth/Downloads/Datasets/test_data.npy",test_data)
```
